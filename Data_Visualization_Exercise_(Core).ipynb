```python
# import libraries  
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Load the dataset 
# Keep in mind that the filename will likely be different on your computer (especially later in course)
filename = '/content/drive/MyDrive/01 COURSE/01 Coding Dojo/01 Colaboratory/04 Week 4/Effective labour market exit age.xlsx'
df = pd.read_excel(filename)
df.head()
```

```python
# Group the data by country and calculate the mean retirement age
avg_age_by_country = df.groupby('Country')['Average Age'].mean().reset_index()
```

```python
# Sort the data by average age
avg_age_by_country = avg_age_by_country.sort_values('Average Age', ascending=False)
```

```python
# Create a horizontal bar chart
plt.barh(avg_age_by_country['Country'], avg_age_by_country['Average Age'])
plt.title('Average Age of Labor Market Exit by Country')
plt.xlabel('Average Age')
plt.ylabel('Country') 
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12,6))

# Create a bar chart
ax.bar(avg_age_by_country['Country'], avg_age_by_country['Average Age'], width=0.5)
ax.set_title('Average Age of Labor Market Exit by Country')
ax.set_xlabel('Country')
ax.set_ylabel('Average Age') 
ax.set_xticklabels(avg_age_by_country['Country'], rotation = 45, ha="right")

# Adjust the spacing between the bars and their labels
plt.subplots_adjust(bottom=0.2)
```
